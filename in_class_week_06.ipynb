```python
# 1c.  How many unique shows were there each year?

# Dropping duplicates by Show Number
no_dup_shows=jeopardy.drop_duplicates(subset ="Show Number")

# Create a list of only the years from the "Air Date" column
Year = []
for yr in no_dup_shows[" Air Date"]:
    year = yr[-4:]
    Year.append(year)
    
# Add the "Year" list as a new column in the jeopardy dataframe
Year_Series=pd.Series(Year, index=no_dup_shows.index)
no_dup_shows = pd.concat([no_dup_shows, Year_Series.rename("Year")], axis=1)

#Sort according to year and show number
no_dup_sorted=no_dup_shows.sort_values(["Year","Show Number"])

#Count the number of occurrences for each year in "Year" column (which is equal to number of shows in each year)
num_of_shows_each_yr=no_dup_sorted["Year"].value_counts()
num_of_shows_each_yr_sorted=num_of_shows_each_yr.sort_index()
pd.DataFrame(num_of_shows_each_yr_sorted)
```

```python
# 1d.  How many times was each category used?  -- I assumed this means the numbers of episodes for which the producers
# decided to use the category.

# Use the dataframe from 1c above with duplicate show rows removed, then count 
cat_ct = no_dup_shows[" Category"].value_counts().head()
pd.DataFrame(cat_ct)
```

```python
# 1e.  Which month (over the years) has had the most shows?  -- The month of November has had the most shows over the years.

#Create a list in order to add a Month column to the no_dup_shows dataframe
month_lst=[]
for i in no_dup_shows[" Air Date"]:
    mo = i[:2]
    month_lst.append(mo)

# Add "month_lst" as a new column in the no_dup_shows dataframe
no_dup_shows["Month"]=month_lst
num_of_shows_by_month=no_dup_shows["Month"].value_counts()
pd.DataFrame(num_of_shows_by_month)
```
